```python
from PIL import Image
import numpy as np
from matplotlib.colors import ListedColormap

from matplotlib import pyplot as plt
```

```python
def create_cityscapes_colormap():
    colormap = np.zeros((256, 3), dtype=np.uint8)
    colormap[:19, :] = np.array([
        [128, 64, 128],
        [244, 35, 232],
        [70, 70, 70],
        [102, 102, 156],
        [190, 153, 153],
        [153, 153, 153],
        [250, 170, 30],
        [220, 220, 0],
        [107, 142, 35],
        [152, 251, 152],
        [70, 130, 180],
        [220, 20, 60],
        [255, 0, 0],
        [0, 0, 142],
        [0, 0, 70],
        [0, 60, 100],
        [0, 80, 100],
        [0, 0, 230],
        [119, 11, 32]])
    return ListedColormap(colormap / 255., N=256)

cmap = create_cityscapes_colormap()
```

```python
id = '00000041'
image = np.array(Image.open(f'images/{id}.jpg'))
label = np.array(Image.open(f'labels/{id}.png'))
```

```python
alpha = 0.5

fig, ax = plt.subplots(3,1, figsize=(18, 30))
for a in ax: a.set_axis_off()
ax[0].imshow(image)
ax[1].imshow(cmap(label))
ax[2].imshow(alpha * image.astype('float') / 255 + (1-alpha) * cmap(label)[..., :3])
plt.savefig('LabelVisualize-41.jpg')

fig.tight_layout()
```
